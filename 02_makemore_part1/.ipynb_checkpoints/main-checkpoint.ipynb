```python
with open("names.txt", "r") as f:
    words = f.read().splitlines()
```

```python
words[:10]
```

```python
len(words)
```

```python
chars = sorted(set("".join(words)))
```

```python
# string to integer
stoi = {c: i+1 for i, c in enumerate(chars)}
stoi["."] = 0

# integer to string
itos = {i: c for c, i in stoi.items()}
```

```python
import torch
N = torch.zeros(27,27, dtype=torch.int32)
# counting loop
for w in words:
    word = "." + w + "."
    for i in range(len(word) -1):
        ix1 = stoi[word[i]]
        ix2 = stoi[word[i+1]]
        N[ix1, ix2] += 1            
```

```python
# Probability matrix
P = (N + 1).float() / (N + 1).sum(dim=1, keepdim=True) #keepdim have relation "broadcasting".
```

```python
# validation
P[0].sum()
```

```python
ix = 0
g = torch.Generator()
g.manual_seed(2147483647) #constant seed for reproducibility
test = []
for i in range(10):
    while True:
        ix = torch.multinomial(P[ix], num_samples=1, generator=g).item()
        if ix == 0:
            print("".join(test))
            test = []
            break
        else:
            test.append(itos[ix])
```

```python
log_likelihood = 0.
n = 0
for w in words:
    word = "." + w + "."
    for i in range(len(word) -1):
        ix1 = stoi[word[i]]
        ix2 = stoi[word[i + 1]]
        n += 1
        log_likelihood += torch.log(P[ix1, ix2])
print(f"Log Likelihood: {log_likelihood}")
print(f"Negative likelihood: {-log_likelihood}")
print(f"Mean of NLL: {-log_likelihood / n}")
```

```python
# Introduction to Neural Networks (dataset preparation)
xs = []
ys = []

for w in words:
    word = "." + w + "."
    for i in range(len(word) - 1):
        xs.append(stoi[word[i]])
        ys.append(stoi[word[i + 1]])
xs = torch.tensor(xs)
ys = torch.tensor(ys)
```

```python
W = torch.randn(27,27, requires_grad=True)

import torch.nn.functional as F
xenc = F.one_hot(xs, num_classes=27).float()
logits = xenc @ W
logits.shape

counts = logits.exp()
probs = counts / counts.sum(1, keepdim=True)
```

```python
loss = -probs[torch.arange(xs.shape[0]), ys].log().mean()
```

```python
W.grad = None
loss.backward()
```

```python
W.data += -0.1 * W.grad
```

```python
xenc = F.one_hot(xs, num_classes=27).float()
logits = xenc @ W
counts = logits.exp()
probs = counts / counts.sum(1, keepdim=True)
loss = -probs[torch.arange(xs.shape[0]), ys].log().mean()
loss
```

```python
W = torch.randn(27, 27, requires_grad=True)

for i in range(5000):

    #forward pass
    xenc = F.one_hot(xs, num_classes=27).float()
    logits = xenc @ W
    counts = logits.exp()
    probs = counts / counts.sum(1, keepdim=True)
    #loss function
    loss = -probs[torch.arange(xs.shape[0]), ys].log().mean()

    W.grad = None
    loss.backward()

    W.data += -0.60 * W.grad

print(loss)
```

```python
g = torch.Generator()
g.manual_seed(2147483647)

isim = []
ix = 0  

for i in range(5):
    while True:
        xenc = F.one_hot(torch.tensor([ix]), num_classes=27).float()
        
        logits = xenc @ W  # (1,27) @ (27,27) = (1,27)
        
        counts = logits.exp()
        probs = counts / counts.sum(1, keepdim=True)  # (1,27)
        
        ix = torch.multinomial(probs[0], num_samples=1, generator=g).item()
        
        if ix == 0:
            print("".join(isim))
            isim = []
            break
        
        isim.append(itos[ix])
```
